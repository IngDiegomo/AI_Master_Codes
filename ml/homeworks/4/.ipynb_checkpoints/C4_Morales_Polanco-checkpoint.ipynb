### c. Divide la base de datos en conjunto de entrenamiento y prueba y crea diversas arquitecturas para el perceptrón multicapa.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

Para facilitar las pruebas de arquitecturas, se utilizará la función "probando" vista en la práctica de clase, y se crearan diversas combinaciones de parámetros con los cuáles ejecutar la función probando. Se modificará la función para omitir la parte de la gráfica en 2D.

```python
def probando(f_act, capas, mi_max_iter=200):
    mlp_prueba = MLPClassifier(solver='lbfgs', random_state=42, 
                               max_iter=mi_max_iter,hidden_layer_sizes=capas,activation=f_act)
    mlp_prueba.fit(X_train, y_train)
    print(f"\nCon la función de activación: {f_act}, capas: {capas} y {mi_max_iter} iteraciones:")
    print("Rendimiento en el conjunto de entrenamiento: ", mlp_prueba.score(X_train, y_train))
    print("Rendimiento en el conjunto de prueba: ",mlp_prueba.score(X_test,y_test))
    print("Rendimiento en el conjunto total: ",mlp_prueba.score(X,y))

    return mlp_prueba.score(X_test,y_test)
    
```

```python
arquitecturas = [['relu', (2,) , 500], ['relu', (3,2) , 500], ['logistic', (10,) , 500], ['tanh', (10,10,) , 500], ['relu', (10,) , 500],
                 ['logistic', (10,39) , 500], ['tanh', (10,10) , 500]]
```

### d. Encuentra las medidas de rendimiento (score) del árbol sobre los conjuntos de entrenamiento, prueba y total.

```python
mejor_score_test = 0
peor_score_test = 1
mejor_arquitectura = 0
peor_arquitectura = 0
for arquitectura in arquitecturas:
    score_test = probando(arquitectura[0], arquitectura[1], arquitectura[2]) 
    if  score_test > mejor_score_test:
        mejor_score_test = score_test
        mejor_arquitectura = arquitectura
    if score_test < peor_score_test:
        peor_score_test = score_test
        peor_arquitectura = arquitectura

print(f"\nMejores resultados encontrados con la función de activación: {mejor_arquitectura[0]}, capas: {mejor_arquitectura[1]} y {mejor_arquitectura[2]} iteraciones:")
print("Rendimiento en el conjunto de prueba: ",mejor_score_test)

print(f"\nPeores resultados encontrados con la función de activación: {peor_arquitectura[0]}, capas: {peor_arquitectura[1]} y {peor_arquitectura[2]} iteraciones:")
print("Rendimiento en el conjunto de prueba: ",peor_score_test)

```

## 3. Al introducir un cambio (optimizador, arquitectura, medida de error, ...) en el modelo que presente una mejora de rendimiento sobre el conjunto de prueba, debes dar una razón por qué crees que el modelo mejora.

Las arquitecturas que probamos tuvieron un rendimiento bastante elevado en todos los conjuntos. Por esto y para completar esta parte de la práctica, se partirá de la arquitectura con peor resultado y se introducirán cambios justificados para mejorar el rendimiento del modelo. Se ejecuta nuevamente la función probando con la peor arquitectura vista en el apartado anterior (función de activación relu con 2 neuronas en una capa intermedia).

```python
arquitectura = ['relu', (2,) , 500]
probando(arquitectura[0], arquitectura[1], arquitectura[2])
```

Una posible fuente de error es el tipo de encoding que se utilizó para las variables de la data. En "A Comparative Study of Categorical Variable Encoding
Techniques for Neural Network Classifier" por Kedar Potdar et al., (https://www.semanticscholar.org/paper/A-Comparative-Study-of-Categorical-Variable-for-Potdar-Pardawala/2609e3c191e284a28cd7c6e9022528ac9b8f4bfc) se puede ver que hay otras maneras de codificar las variables categóricas que pueden llevar a mejores resultados. 

La razón por la cuál el label encodingo o "ordinal encoding", el cuál es el encoding utilizado hasta ahora en la tarea puede estar afectando el rendimiento del modelo es porque esta codificación introduce un orden a las categorias que puede que no exista realmente. Una alternativa a este encoding es el encoding one hot, el cuál es un tipo de codificación categórica que no asume un orden en las categorías sino que añade una columna por cada categoría de la variable, la cuál tendra un 1 o un 0.

Se puede hacer el encoding one hot utilizando la función dummies de pandas

```python
# Usando One hot encoding para X
X_one_hot = df.drop('class', axis=1)
X_one_hot = pd.get_dummies(X_one_hot, columns=X_one_hot.columns)

# Haciendo lo mismo con y
y_one_hot = df['class']
y_one_hot = pd.get_dummies(y_one_hot, prefix='class')

# Las matrices resultantes son de estos tamaños
print(X_one_hot.shape)
print(y_one_hot.shape)
```

Es importante recalcar que el cambio en el encoding es a su vez un cambio en la arquitectura de la red, ya que en un principio se tenía un vector de entrada en ${\mathbb R}^{5}$ y un vector de salida en ${\mathbb R}^{1}$, ahora el vector de entrada pertenece a ${\mathbb R}^{21}$ y el de salida a ${\mathbb R}^{4}$. Se procede a comprobar el efecto del encoding en el rendimiento de la peor arquitectura.

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y_one_hot, stratify = y_one_hot, random_state=42)
```

```python
# Recordando la arquitectura con peor desempeño de la sección anterior
arquitectura = ['relu', (2,) , 500]

mlp_prueba = MLPClassifier(solver='lbfgs', random_state=42, max_iter=arquitectura[2], hidden_layer_sizes=arquitectura[1], activation=arquitectura[0])
mlp_prueba.fit(X_train, y_train)

print(f"\nCon la función de activación: {arquitectura[0]}, capas: {arquitectura[1]} y {arquitectura[2]} iteraciones:")

print("Rendimiento en el conjunto de entrenamiento: ", mlp_prueba.score(X_train, y_train))
print("Rendimiento en el conjunto de prueba: ",mlp_prueba.score(X_test,y_test))
print("Rendimiento en el conjunto total: ",mlp_prueba.score(X_one_hot,y_one_hot))
```

Se puede apreciar un aumento en el rendimiento de la arquitectura sobre el conjunto de prueba de aproximadamente un 15%, manteniendo una arquitectura bastante simple de solo 1 capa intermedia y 2 neuronas. Se especula que la razón de este aumento es debido a que el label encoding introducía un orden numérico inexistente en las variables. También, el aumento de dimensionalidad en la entrada y en la salida pudo jugar un papel importante para que la red encontrara una mejor separación entre las clases.

Otro cambio que se puede introducir es a la función de activación. Recordando la forma de la función ReLu y lo visto en clases, esta separa el espacio en lineas rectas, tal vez esto no es lo suficientemente expresivo para el problema actual. Por esto se introduce el cambio de utilizar la función sigmoide como función de activación, la cual puede hacer separaciones más complejas del espacio. Se procede a comprobar el efecto de la función sigmoide en el rendimiento de la arquitectura con peor desempeño de la sección anterior, junto con el one hot encoding.

```python
# Recordando la arquitectura con peor desempeño de la sección anterior
arquitectura = ['logistic', (2,) , 500]

mlp_prueba = MLPClassifier(solver='lbfgs', random_state=42, max_iter=arquitectura[2], hidden_layer_sizes=arquitectura[1], activation=arquitectura[0])
mlp_prueba.fit(X_train, y_train)

print(f"\nCon la función de activación: {arquitectura[0]}, capas: {arquitectura[1]} y {arquitectura[2]} iteraciones:")

print("Rendimiento en el conjunto de entrenamiento: ", mlp_prueba.score(X_train, y_train))
print("Rendimiento en el conjunto de prueba: ",mlp_prueba.score(X_test,y_test))
print("Rendimiento en el conjunto total: ",mlp_prueba.score(X_one_hot,y_one_hot))
```

Utilizar la función sigmoide no mejoró el rendimiento del modelo, de hecho lo empeoró. Esto resulta anti intutivo debido a que luego de hacer one hot encoding, se está trabajando con valores binarios, justo el tipo de valores que esta función aproxima. Tal vez al ser un problema multiclase, aunque las variables sean binarias la sigmoide funciona peor que la ReLu en este caso.

Al inicio de esta práctica se observó que había un desbalance en las clases:

```python
# Número de clases únicas en y
y = df['class']
print("Número de clases únicas en y:")
print(y.nunique())

# Cuenta de cada clase de y
print("\nCuenta de cada clase de y:")
print(y.value_counts())
```

Se está tomando esto en cuenta con el parámetro "stratify" al hacer los splits de train y test, esto asegura que cada conjunto de train y test tenga una proporción de elementos de cada clase parecida a la proporción total del dataset.

Una última modificación que se puede hacer es aumentar el número de neuronas de la capa intermedia. La razón por la que se espera que esto mejore el rendimiento es porque la dimensionalidad de la capa de entrada y de salida aumento, por lo que el espacio se volvió más complejo y puede que el modelo necesite una mayor expresividad para separarlo satisfactoriamente. Se probaran 10 neuronas.

```python
# Recordando la arquitectura con peor desempeño de la sección anterior
arquitectura = ['relu', (10,) , 500]

mlp_prueba = MLPClassifier(solver='lbfgs', random_state=42, max_iter=arquitectura[2], hidden_layer_sizes=arquitectura[1], activation=arquitectura[0])
mlp_prueba.fit(X_train, y_train)

print(f"\nCon la función de activación: {arquitectura[0]}, capas: {arquitectura[1]} y {arquitectura[2]} iteraciones:")

print("Rendimiento en el conjunto de entrenamiento: ", mlp_prueba.score(X_train, y_train))
print("Rendimiento en el conjunto de prueba: ",mlp_prueba.score(X_test,y_test))
print("Rendimiento en el conjunto total: ",mlp_prueba.score(X_one_hot,y_one_hot))
```

Se puede que el solver falla en converger en el número de iteraciones colocado. Pero también hay una pequeña mejora en el rendimiento del modelo al utilizar 10 neuronas en lugar de 2 en la capa de entrada. Finalmente se probará con un número mayor de iteraciones para observar si la convergencia del solver resulta en un mejor resultado.

```python
# Recordando la arquitectura con peor desempeño de la sección anterior
arquitectura = ['relu', (10,) , 2000]

mlp_prueba = MLPClassifier(solver='lbfgs', random_state=42, max_iter=arquitectura[2], hidden_layer_sizes=arquitectura[1], activation=arquitectura[0])
mlp_prueba.fit(X_train, y_train)

print(f"\nCon la función de activación: {arquitectura[0]}, capas: {arquitectura[1]} y {arquitectura[2]} iteraciones:")

print("Rendimiento en el conjunto de entrenamiento: ", mlp_prueba.score(X_train, y_train))
print("Rendimiento en el conjunto de prueba: ",mlp_prueba.score(X_test,y_test))
print("Rendimiento en el conjunto total: ",mlp_prueba.score(X_one_hot,y_one_hot))
```

La convergencia del solver resultó en un rendimiento ligeramente mejor en el conjunto de entrenamiento pero un rendimiento ligéramente peor en el conjunto de prueba, por lo que el modelo hizo overfitting. El resultado anterior con 500 iteraciones es preferible.
